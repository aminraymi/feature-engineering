```python
df['time_part'] = df['date'].dt.time
df['time_part'].head()
```

```python
rng_ = pd.date_range('2019-03-05', periods=20, freq='M')
df = pd.DataFrame({'date': rng_})
df.head()
```

```python
df['year'] = df['date'].dt.year
df.head()
```

```python
df['month'] = df['date'].dt.month
df.head()
```

```python
df['quarter'] = df['date'].dt.quarter
df.head()
```

```python
df['semester'] = np.where(df['quarter'].isin([1,2]), 1, 2)
df.head()
```

```python
rng_ = pd.date_range('2019-03-05', periods=20, freq='D')
df = pd.DataFrame({'date': rng_})
df.head()
```

```python
df['day_mo'] = df['date'].dt.day
df.head()
```

```python
df['day_week'] = df['date'].dt.dayofweek
df.head()
```

```python
df['day_week_name'] = df['date'].dt.day_name()
df.head()
```

```python
df['is_weekend'] = np.where(df['day_week_name'].isin(['Sunday',
'Saturday']), 1, 0)
df.head()
```

```python
df['week'] = df['date'].dt.isocalendar().week
df.head()
```

```python
rng_ = pd.date_range('2019-03-05', periods=20, freq='1h15min10s')
df = pd.DataFrame({'date': rng_})
df.head()
```

```python
df['hour'] = df['date'].dt.hour
df['min'] = df['date'].dt.minute
df['sec'] = df['date'].dt.second
df.head()
```

```python
df[['h','m','s']] = pd.DataFrame([(x.hour, x.minute, x.second) for
x in df['date']])
df.head()
```

```python
df['is_morning'] = np.where( (df['hour'] < 12) & (df['hour'] > 6),
1, 0 )
df.head()
```
