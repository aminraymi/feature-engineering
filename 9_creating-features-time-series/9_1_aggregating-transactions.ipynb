```python
# In this recipe, we will create a flattened dataset by aggregating multiple transactions using
# common mathematical operations. We will use pandas to do this
%config Completer.use_jedi = False
import pandas as pd
import featuretools as ft
```

```python
data_dict = ft.demo.load_mock_customer()
```

```python
data = data_dict["transactions"].merge(data_dict["sessions"]).merge(data_dict["customers"])
```

```python
data.head()
```

```python
data = data[['customer_id', 'transaction_id', 'transaction_time','amount']]
```

```python
data.head()
```

```python
operations = ['sum', 'max', 'min', 'mean', 'median', 'std','count']
```

```python
feature_names = [ 'total_amount', 'max_amount', 'min_amount',
'mean_amount', 'median_amount', 'std_amount',
'number of transactions']
```

```python
df = pd.DataFrame()
df[feature_names] = data.groupby('customer_id')['amount'].agg(operations)
df
```
